```python
# To Create 500 random numbers with a two-dimensional normal distribution

import numpy as np
import matplotlib.pyplot as plt

mean = (-3, 0)
cov_mx = np.array([[1, .8], [.8, 1]])

rand = np.random.multivariate_normal(mean, cov_mx, (500,))

np.random.seed(0)
print(rand.shape)
     
```

```python
np.random.seed(0)
```

```python
#Visualization by scatter plot

x, y = np.hsplit(rand, 2)

plt.scatter(x, y)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot of random data")
plt.show()
```

```python
#Visualization by histogram

plt.hist(x, 50)

plt.xlim((x.min(), x.max()))

plt.xlabel("X")
plt.ylabel("Frequency")
plt.title("Histogram x1")
plt.show()
```

```python


plt.hist(y, 50)

plt.xlim((y.min(), y.max()))

plt.xlabel("Y")
plt.ylabel("Frequency")
plt.title("Histogram x2")
plt.show()
```

```python
#visualization of the data of problems 1 and 4 as one scatter plot

mean = (0, -3)
cov_mx = np.array([[1, .8], [.8, 1]])

rand1 = np.random.multivariate_normal(mean, cov_mx, 500)

np.random.seed(1)

x1, y1 = np.hsplit(rand1, 2)

plt.scatter(x, y)

plt.scatter(x1, y1, color="r")
plt.legend([0, 1])

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")
plt.show()
```

```python
#Data combination

bundle = np.vstack((rand, rand1))

bundle.shape
```
