```python
#Linear function

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-50, 50.1, 0.1)

def linear(x):
  return (.5 * x + 1)

y = linear(x)

x.shape
```

```python
#Combining arrays

x = x.reshape((1001, 1))
y = y.reshape((1001, 1))

z = np.concatenate((x, y), 1)

z.shape
```

```python
difference = np.diff(z, 1, 0)

gradient = difference.T[1] / difference.T[0]

print(gradient)
```

```python
plt.plot(x[1:], gradient)
plt.xlabel("X")
plt.ylabel("Gradient")

plt.title("A graph of X against Y")
plt.show()
```

```python
def square(x):
  return x**2

def sin(x):
  operand = (x**0.5)
  return np.sin(operand)

def func(x):
  return ((2*x**2) + 2**x)


def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
  x = np.arange(x_range[0], x_range[1], x_range[2])
  size = x.size
  x = x.reshape((size, 1))
  y = function.reshape((size, 1))
  z = np.concatenate((x, y), 1)

  difference = np.diff(z, 1, 0)
  gradient = difference.T[1] / difference.T[0]

  return z, gradient
```

```python
x = np.arange(-50, 50.1, 0.1)

a, b = compute_gradient(sin(x))

x_axis = a.T[0]
y_axis = a.T[1]
plt.plot(x_axis, y_axis)
plt.show()
```

```python
x = np.arange(-50, 50.1, 0.1)

a, b = compute_gradient(func(x))

x_axis = a.T[0]
y_axis = a.T[1]
plt.plot(x_axis, y_axis)
plt.show()
```
